## **Partie 1 — Programmation linéaire avec PuLP**

```python
from pulp import *

# a) Définir le problème (maximiser)
prob = LpProblem("Partie1Pulp", LpMaximize)

# b) Créer les variables de décision (avec bornes)
joueurs = [
  {"nom": "Alice", "score": 88, "salaire": 1200, "poids": 72},
  {"nom": "Bob", "score": 91, "salaire": 1800, "poids": 85},
  {"nom": "Clara", "score": 84, "salaire": 950, "poids": 68},
  {"nom": "David", "score": 93, "salaire": 2100, "poids": 90},
  {"nom": "Emma", "score": 79, "salaire": 800, "poids": 65},
  {"nom": "Frank", "score": 87, "salaire": 2400, "poids": 95},
  {"nom": "Grace", "score": 85, "salaire": 1050, "poids": 70},
  {"nom": "Hugo", "score": 89, "salaire": 1600, "poids": 80},
]

equipes = ["A", "B"]

joueurs_individuels = len(joueurs)

score_AB = 0
salaire_AB = 0
poids_AB = 0

x = LpVariable.dicts("x", (range(joueurs_individuels), equipes), cat=LpBinary)


# c) Ajouter les contraintes
for e in equipes:
  prob += lpSum(x[i][e] for i in range(joueurs_individuels)) == 3

for i in range(joueurs_individuels):
  prob += lpSum(x[i][e] for e in equipes) <= 1

for e in equipes:
  prob += lpSum(joueurs[i]["salaire"] *x[i][e] for i in range(joueurs_individuels) for e in equipes) <= 8500

for e in equipes:
  prob += lpSum(joueurs[i]["poids"] * x[i][e] for i in range(joueurs_individuels)) <= 250

# d) Définir la fonction objective
prob += lpSum(joueurs[i]["score"] * x[i][e] for i in range(joueurs_individuels) for e in equipes)

# e) Résoudre
prob.solve()
print("Status:", LpStatus[prob.status])

# f) Afficher la valeur optimale de Z
print(f"Z optimal : {prob.objective.value()}")

for e in equipes:
  print("Equipe: ", e)

  joueursdans_equipes = [joueurs[i] for i in range(joueurs_individuels) if value(x[i][e]) == 1]

  for b in joueursdans_equipes:
    print(b["nom"])

  score_total = sum(b["score"] for b in joueursdans_equipes)
  salaire_total = sum(b["salaire"] for b in joueursdans_equipes)
  poids_total = sum(b["poids"] for b in joueursdans_equipes)

  print("Score total: ", score_total)
  print("Salaire total: ", salaire_total)
  print("Poids total: ", poids_total)

  score_AB += score_total
  salaire_AB += salaire_total
  poids_AB += poids_total



print("Totaux des 2 equipes")
print("Score total: ", score_AB)
print("Salaire total: ", salaire_AB)
print("Poids total: ", poids_AB)

```

```python
from itertools import combinations

# Definir les joueurs
joueurs = [
  {"nom": "Alice", "score": 88, "salaire": 1200, "poids": 72},
  {"nom": "Bob", "score": 91, "salaire": 1800, "poids": 85},
  {"nom": "Clara", "score": 84, "salaire": 950, "poids": 68},
  {"nom": "David", "score": 93, "salaire": 2100, "poids": 90},
  {"nom": "Emma", "score": 79, "salaire": 800, "poids": 65},
  {"nom": "Frank", "score": 87, "salaire": 2400, "poids": 95},
  {"nom": "Grace", "score": 85, "salaire": 1050, "poids": 70},
  {"nom": "Hugo", "score": 89, "salaire": 1600, "poids": 80},
]

s1_equipeA = []
s1_equipeB = []



# triez la liste joueurs par score décroissant
trier_joueurs = sorted(joueurs, key=lambda j: j["score"], reverse=True)
# meilleur score absolu
for b in trier_joueurs:
  salaire_A = sum(b["salaire"] for b in s1_equipeA)
  poids_A = sum(b["poids"] for b in s1_equipeA)

  if len(s1_equipeA) < 3 and salaire_A + b["salaire"] <= 8500 and poids_A + b["poids"] <= 250:
    s1_equipeA.append(b)

for b in trier_joueurs:
  if b in s1_equipeA:
    continue

  salaire_B = sum(b["salaire"] for b in s1_equipeB)
  poids_B = sum(b["poids"] for b in s1_equipeB)

  if len(s1_equipeB) < 3 and salaire_B + b["salaire"] <= 8500 and poids_B + b["poids"] <= 250:
    s1_equipeB.append(b)


joueurs_restants = [b for b in joueurs if b not in equipe_A and b not in s1_equipeB]

print("Equipe A: ", [b["nom"] for b in s1_equipeA])
print("Equipe B: ", [b["nom"] for b in s1_equipeB])

salaire_A = sum(b["salaire"] for b in s1_equipeA)
poids_A = sum(b["poids"] for b in s1_equipeA)

salaire_B = sum(b["salaire"] for b in s1_equipeB)
poids_B = sum(b["poids"] for b in s1_equipeB)

for b in joueurs_restants:
  if salaire_A + b["salaire"] > 8500:
    print(b["nom"], "exlus de l'equipe A car son salaire est trop eleve")
  if salaire_B + b["salaire"] > 8500:
    print(b["nom"], "exclus de l'equipe B car son salaire est trop eleve")
  if poids_A + b["poids"] > 250:
    print(b["nom"], "exclus de l'equipe A car son poids est trop eleve")
  if poids_B + b["poids"] > 250:
    print(b["nom"], "exclus de l'equipe B car son poids est trop eleve")
```
